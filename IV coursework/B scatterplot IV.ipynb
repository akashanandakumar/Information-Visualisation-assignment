```python
#import necessary libraries
import altair as alt
import pandas as pd
import vega_datasets
```

```python
#create dataframe using pandas 
df = pd.read_csv('/Users/prathiksha/Downloads/pokemon_adnan_cleaned.csv')
df.head(5)
```

```python
# create a dictionary to map old tiers to new tiers
tier_mapping = {'OUBL': 'Uber', 'UUBL': 'OU', 'RUBL': 'UU', 'NUBL': 'RU', 'PUBL': 'NU'}

# use the map() method to replace the old tiers with new tiers based on the dictionary
df['Tier'] = df['Tier'].map(tier_mapping).fillna(df['Tier'])

# filter the dataframe to include only the desired tiers
allowed_tiers = ['Uber', 'OU', 'UU', 'RU', 'NU', 'PU']
df = df[df['Tier'].isin(allowed_tiers)]
```

```python
#use only the required columns
columns_to_keep = ['Name','Tier','Base Stat Total','Num Moves']
df = df[columns_to_keep]
```

```python
#create scatter plot for the data

alt.Chart(df).mark_point().encode(
    x='Base Stat Total',
    y='Num Moves',
    color='Tier',
    tooltip='Name'
).properties(
    title='Pokemon Data'
).interactive()
```

```python
#create scatter plot and bar graph using brushing and linking to make it interative

multi = alt.selection_multi(fields=['Tier'], empty='all')
interval = alt.selection_interval(encodings=['y'])

scatter = alt.Chart(df).mark_point().encode(
    x='Base Stat Total',
    y='Num Moves',
    color=alt.condition(multi & interval, 'Tier', alt.value('lightgray'))
).properties(
    selection=interval
)

bar = alt.Chart(df).mark_bar().encode(
    x='average(Base Stat Total)',
    y='Tier',
    color=alt.condition(multi, 'Tier', alt.value('lightgray'))
).properties(
    selection=multi
).transform_filter(
    interval
)

plot = scatter & bar
plot
```
